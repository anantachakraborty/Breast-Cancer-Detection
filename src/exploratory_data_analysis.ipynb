Correlation Matrix (heatmap) visualizing

```python
# Load the breast cancer dataset
data = load_breast_cancer()

# Create a DataFrame from the dataset
df = pd.DataFrame(data.data, columns=data.feature_names)

# Group features by categories
feature_categories = {
    "Mean Features": ["mean radius", "mean texture", "mean perimeter"],
    "Error Features": ["radius error", "texture error"],

}

# Create correlation matrices and heatmaps for each feature
for category, features in feature_categories.items():
    category_df = df[features]
    correlation_matrix = category_df.corr()

    plt.figure(figsize=(8, 6))
    sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
    plt.title(f"Correlation Matrix - {category}")
    plt.show()
```

```python
plt.figure(figsize=(8, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
